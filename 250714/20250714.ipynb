```python
def my_sum(x):
    result=[]
    for i in x:
        if i % 2 == 0:
            result.append(i)
    return sum(result)
```

```python
x = list(range(1,11))

my_sum(x)
```

```python
def get_even(x):
    for i in x:
        if i % 2 != 0:
            x.remove(i)
```

```python
x = list(range(1,11))

get_even(x.copy())
sum(x)
```
