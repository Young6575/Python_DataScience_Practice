```python
from sklearn.datasets import load_iris
```

1. 데이터 -> irsi
2. Pandas DataFrame으로 만들어야 함.

```python
iris = load_iris()
x = iris.data
y = iris.target

feature_names = iris.feature_names
```

```python
iris.data[1]
```

```python
df = pd.DataFrame(x, columns=feature_names)
df["species"] = [str(iris.target_names[i]) for i in y]
df.corr(numeric_only=True)
```

```python
# petal length (cm)가 1.5 보다 큰 데이터를 출력을 해보자!
mask = df["petal length (cm)"] > 1.5 # Bool daa frame => 필터링
df_petal = df[mask]

df_petal
```

```python
ax = df_petal.plot()
ax.set_title("붓꽃 데이터")
ax.set_xlabel("개별 데이터")
ax.set_ylabel("관측 데이터")
```

```python
means = df.groupby("species").mean()
colors = ["red", "blue", "green"]
species_list = iris.target_names
```

```python
plt.figure(figsize=(12,8))
for i, species in enumerate(species_list):
    print(i, species)
    plt.plot(iris.feature_names, means.loc[species],
            color=colors[i], label=species)



plt.title("종별 특성 평균값 비교")
plt.xlabel("특성")
plt.ylabel("평균값 (cm)")
plt.legend()
plt.show()
```
