```python
def make_df(cols, ind):
    """빠르게 DataFrame 생성"""
    data = {c: [str(c) + str(i) for i in ind]
                for c in cols}
    return pd.DataFrame(data, ind)


# DataFrame 예제
make_df('ABC', range(3))
```

```python
abbrevs = pd.read_csv('https://raw.githubusercontent.com/jakevdp/data-USstates/refs/heads/master/state-abbrevs.csv')
areas = pd.read_csv('https://raw.githubusercontent.com/jakevdp/data-USstates/refs/heads/master/state-areas.csv')
pop = pd.read_csv('https://raw.githubusercontent.com/jakevdp/data-USstates/refs/heads/master/state-population.csv')

print(abbrevs.head()); print(abbrevs.info());
```

```python
areas.head()
```

```python
pop.head()
```

```python
merged = pd.merge(pop,abbrevs,how='outer',left_on='state/region', right_on='abbreviation')

merged
```

```python
merged = merged.drop('abbreviation', 1) # 중복 정보 삭제
merged.head()
```

```python
merged.isnull().any()
```

```python
merged[merged['population'].isnull()].head()
```

```python
merged.loc[merged['state/region'] == 'PR', 'state'] = 'Puerto Rico'
merged.loc[merged['state/region'] == 'USA', 'state'] = 'United States'
```

```python
final = pd.merge(merged, areas,on='state', how='left')
```

```python
final.isnull().any()
```

```python
final['state'][final['area (sq. mi)'].isnull()].unique()
```

```python
data2010 = final.query("year == 2010 & ages == 'total'")
data2010.head()
```

```python
data2010.set_index('state', inplace=True)
density = data2010['population'] / data2010['area (sq. mi)']
```

```python
density.sort_values(ascending=False, inplace=True)
density.head()
```
