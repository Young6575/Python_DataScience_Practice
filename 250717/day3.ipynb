방법 1. None으로 표기

```python
vals1 = np.array([1,None,2,3])
vals1
```

```python
%timeit np.arange(1E6, dtype=int).sum()
```

```python
%timeit np.arange(1E6, dtype=object).sum() # dtype=object는 파이썬 객체임을 의미. 연산이 느림
```

```python
vals1.sum() # None이 있기 때문에 sum,min같은 집계 함수는 오류 발생
```
